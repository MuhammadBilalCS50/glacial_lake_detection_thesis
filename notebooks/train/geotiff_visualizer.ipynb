```python
import rasterio
import rasterio.plot
from pathlib import Path
import numpy as np

bands=(6,4,2)

def read_geotiff(path: Path):
    with rasterio.open(path) as src:
        arr = src.read()  # (C,H,W)
    return arr

def select_rgbnir(x: np.ndarray, bands=(6,4,2)):
    idx = [b-1 for b in bands]
    return x[idx, ...].astype(np.float32)

data_name = Path("D:/Thesis/glacial_lake_detection_thesis/Training/3 Preparing train data/rgb_outputs/chips_out/20200923_050442_18_222c_3B_Visual_clip_reproject_file_format/20200923_050442_18_222c_3B_Visual_clip_reproject_file_format_r0_c818_512px.tif")
tiff8 = read_geotiff(data_name)
#tiff = select_rgbnir(tiff8, bands)
rasterio.plot.show(tiff8, title = "HKH")
```

```python
import pandas as pd
from pathlib import Path
import shutil

excel = "D:/Thesis/glacial_lake_detection_thesis/Training/3 Preparing train data/8_band_outputs/subset_800/paths.xlsx"          # <-- change
sheet = "Sheet1"              # <-- change or None
column = "label_path"               # <-- change if needed
dest = Path("D:/Thesis/glacial_lake_detection_thesis/Training/3 Preparing train data/8_band_outputs/subset_800/labels")    # <-- change

df = pd.read_excel(excel, sheet_name=sheet)  # or read_csv for CSV
dest.mkdir(parents=True, exist_ok=True)

for p in df[column].dropna().astype(str):
    src = Path(p).expanduser()
    if not src.is_absolute():
        src = Path.cwd() / src
    if src.exists() and src.is_file():
        shutil.copy2(src, dest / src.name)
```
